```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

***Reading a Sentiment File***

```python
df = pd.read_csv('sentiment_dataset.csv')
```

```python
print("Dataset Preview:")
print(df.head(20))
```

***Calculating number of positive and negative counts***

```python
positive_count = df['Sentiment'].sum()
```

```python
negative_count=len(df)-positive_count
```

```python
print("Positive count: ",positive_count)
print("Negative count: ",negative_count)
```

***Ploting a Bar Graph***

```python
plt.figure(figsize=(6, 4))
plt.bar(["Positive", "Negative"], [positive_count, negative_count], color=['blue', 'red'])
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.title("Sentiment Distribution in Dataset")
plt.show()
```

***Calculating Percentage of positive and negative feedback...***

```python
positive_percentage = (positive_count / len(df)) * 100
negative_percentage = (negative_count / len(df)) * 100
print("Percentage of Sentiments...\n")
print("Postive percentage: ",positive_percentage,"%")
print("Negative percentage: ",negative_percentage,"%")
```
